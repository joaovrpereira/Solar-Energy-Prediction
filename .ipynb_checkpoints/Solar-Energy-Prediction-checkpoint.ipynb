```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

df_gen01 = pd.read_csv('solar-power-generation-data/Plant_1_Generation_Data.csv')
df_weather01 = pd.read_csv('solar-power-generation-data/Plant_1_Weather_Sensor_Data.csv')

df_gen02 = pd.read_csv('solar-power-generation-data/Plant_2_Generation_Data.csv')
df_weather02 = pd.read_csv('solar-power-generation-data/Plant_2_Weather_Sensor_Data.csv')
```

```python
df_gen01['DATE_TIME'] = pd.to_datetime(df_gen01['DATE_TIME'], dayfirst=True)
df_gen02['DATE_TIME'] = pd.to_datetime(df_gen02['DATE_TIME'], dayfirst=True)
```

```python
df_weather01['DATE_TIME'] = pd.to_datetime(df_weather01['DATE_TIME'])
df_weather02['DATE_TIME'] = pd.to_datetime(df_weather02['DATE_TIME'])
```

```python
df_gen01_grouped = (
    df_gen01.groupby(['PLANT_ID', 'DATE_TIME'], as_index=False)
    [['AC_POWER', 'DAILY_YIELD', 'TOTAL_YIELD']]
    .sum()
    )
```

```python
df_gen02_grouped = (
    df_gen02.groupby(['PLANT_ID', 'DATE_TIME'], as_index=False)
    [['AC_POWER', 'DAILY_YIELD', 'TOTAL_YIELD']]
    .sum()
    )
```

```python
df_gen01_grouped['PLANT_ID'] = df_gen01_grouped['PLANT_ID'].astype(str)
df_weather01['PLANT_ID'] = df_weather01['PLANT_ID'].astype(str)
```

```python
df_final_01 = pd.merge(
    df_gen01_grouped,
    df_weather01[['PLANT_ID','DATE_TIME','AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION']],
    on=['PLANT_ID','DATE_TIME'],
    how='inner'
)
```

```python
df_final_01.head()
```

```python
df_gen02_grouped['PLANT_ID'] = df_gen02_grouped['PLANT_ID'].astype(str)
df_weather02['PLANT_ID'] = df_weather02['PLANT_ID'].astype(str)
```

```python
df_final_02 = pd.merge(
    df_gen02_grouped,
    df_weather02[['PLANT_ID','DATE_TIME','AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION']],
    on=['PLANT_ID','DATE_TIME'],
    how='inner'
)
```

```python
df_final_02.nlargest(15, 'AC_POWER')
```

```python
# Estatísticas descritivas básicas

df_final_02.describe()
```

```python
# Mínimo 
## Atributos de data com .dt
df_final_02['DATE_TIME'] = pd.to_datetime(df_final_02['DATE_TIME'])
df_final_02['DATE'] = df_final_02['DATE_TIME'].dt.date

df_grouped = (
    df_final_02.groupby(['PLANT_ID', 'DATE'], as_index=False)
    [['AC_POWER', 'DAILY_YIELD', 'TOTAL_YIELD']]
    .sum()
)

minDay = df_grouped.loc[df_grouped['AC_POWER'].idxmin()]
minDay
```

```python
# Máximo
## Atributos de data com .dt
df_final_02['DATE_TIME'] = pd.to_datetime(df_final_02['DATE_TIME'])
df_final_02['DATE'] = df_final_02['DATE_TIME'].dt.date

df_grouped = (
    df_final_02.groupby(['PLANT_ID', 'DATE'], as_index=False)
    [['AC_POWER', 'DAILY_YIELD']]
    .sum()
)

maxDay = df_grouped.loc[df_grouped['AC_POWER'].idxmax()]
maxDay_df = pd.DataFrame([maxDay]).style.relabel_index(["Val. máximo"])
display(maxDay_df)
```

```python
df_grouped_final = (
    df_final_02.groupby(['PLANT_ID', 'DATE'], as_index=False)
    .agg({
        'AC_POWER': 'sum',
        'AMBIENT_TEMPERATURE': 'mean',
        'IRRADIATION': 'mean',
        'DAILY_YIELD': 'max'
    })
)
display(df_grouped_final)
```

```python
corr_matrix_01 = df_grouped_final.drop(columns=['DATE', 'PLANT_ID']).corr(numeric_only=True)
display(corr_matrix_01)
```

```python
sns.heatmap(corr_matrix_01, cmap="YlGnBu", annot=True)
```
